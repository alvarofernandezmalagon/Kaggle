## Meaning of the values

count: shows how many rows have non-missing values

unique: the number of unique values

top: the most common value

freq: the frequence of the top value

mean: the average

std: measures how numerically spread out the values are

min: the smallest value

25%: first percentile

50% or median: second percentile

75%: third percentile

max: the biggest value

NaN: returned for numeric operations in colums with strings

#### What is the average lot size (rounded to nearest integer)?

```python
avg_lot_size = round(melbourne_data['BuildingArea'].mean())
avg_lot_size
```

#### As of today, how old is the newest home (current year - the date in which it was built)

```python
import datetime
newest_home_age = datetime.date.today().year - melbourne_data[['YearBuilt']].max()
newest_home_age
```

### I can see that the data is updated because there is only one year difference between today and the day of the last house built
